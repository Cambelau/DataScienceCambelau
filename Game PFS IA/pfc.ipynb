## Initiate Logic Game

```python
import random

global resultatLabel,scoreJoueur,scoreOrdinateur

scoreJoueur=scoreOrdinateur=0

# Appuyer sur un bouton déclenche 
def play(movePlayer):  
    listeCoupsJoueur.append(movePlayer)
    moveComputer=getMoveComputer()
    resultat = checkWinner(movePlayer,moveComputer)
    updateResultat(movePlayer,moveComputer,resultat)

#Vérifier qui gagne (move joueur,move ordi) -> -1 égalité, 1 victoire joueur, 0 victoire ordi
def checkWinner(moveP,moveC):
    moves=[moveP,moveC]

    if(moveP==moveC):
        return -1
    #coups gagnants joueur
    if(moves==[2,1] or moves==[1,0] or moves==[0,2]):
        return 1

    return 0

#maj des scores
def updateResultat(moveP,moveC,w):  
    global resultatLabel,scoreJoueurLabel,scoreOrdinateurLabel,scoreJoueur,scoreOrdinateur

    resultatLabel["text"] ="Vous avez joué : ",coups[moveP]," et l'ordi : ",coups[moveC]

    if(w==-1):
        resultatLabel["bg"]="blue"
        return
    if(w==1):
        scoreJoueur=scoreJoueur+1
        resultatLabel["bg"]="green"
        scoreJoueurLabel["text"]=scoreJoueur
    if(w==0):
        scoreOrdinateur=scoreOrdinateur+1
        resultatLabel["bg"]="red"
        scoreOrdinateurLabel["text"]=scoreOrdinateur  

def resetGame():
    global resultatLabel,scoreJoueurLabel,scoreOrdinateurLabel,scoreJoueur,scoreOrdinateur

    resultatLabel["text"] =""
    scoreJoueurLabel["text"]="0"
    scoreOrdinateurLabel["text"]="0"
    scoreJoueur=0
    scoreOrdinateur=0
    listeCoupsJoueur=[]
    listeCoupsOrdi=[]
    listeScoreJoueur=[]
    listeScoreOrdi=[]
```

## Initiate Tkinter window

```python
from tkinter import *
import tkinter as tk
import customtkinter
from PIL import Image
import requests

# url="http://image.over-blog.com/vLpqm2GBbVNxLNOM-eql4v-Jupc=/filters:no_upscale()/image%2F0931882%2F20210329%2Fob_563bff_quitter.png"


class Root(Tk):
    def __init__(self):
        super(Root,self).__init__()
 
        self.title("Python Pierre Feuille Ciseau")
        self.minsize(333,333)
        self.focus()
        self.eval('tk::PlaceWindow . center')

        global resultatLabel,scoreJoueurLabel,scoreOrdinateurLabel,scoreJoueur,scoreOrdinateur

        quitterButton = tk.Button(
            text='Quitter',
            command=self.destroy,
            width=10,
            bg="gray",
            font=("Arial", 8),
            # image=imageQuit,
        )
        resetButton = tk.Button(
            text='Reset',
            command=lambda :resetGame(),
            width=10,
            bg="gray",
            font=("Arial", 8),
        )
        titreLabel = tk.Label(
            text="Jouons au Pierre Feuille Ciseau !",
            height=4,
            font=("Arial", 16, "italic"),
            foreground="blue", 
        )
        joueurLabel = tk.Label(
                text="Score Joueur ",
                height=2,
                font=("Arial", 10, "italic"),
                foreground="green", 
        )
        scoreJoueurLabel = tk.Label(
                text="0",
                height=2,
                font=("Arial", 10, "italic"),
                foreground="green", 
        )
        ordinateurLabel = tk.Label(
                text="Score Ordinateur ",
                height=2,
                font=("Arial", 10, "italic"),
                foreground="red", 
        )
        scoreOrdinateurLabel = tk.Label(
                text="0",
                height=2,
                font=("Arial", 10, "italic"),
                foreground="red", 
        )
        pierreButton= tk.Button(
            text='Pierre',
            command=lambda :play(0),
            width=10,
            bg="white",
            font=("Arial", 8),
        )
        feuilleButton= tk.Button(
            text='Feuille',
            command=lambda :play(1),
            width=10,
            bg="white",
            font=("Arial", 8),
        )
        ciseauxButton= tk.Button(
            text='Ciseaux',
            command=lambda :play(2),
            width=10,
            bg="white",
            font=("Arial", 8),
        )
        resultatLabel = tk.Label(
            text="",
            height=2,
            font=("Arial", 12, "italic"),
            foreground="black", 
            # width=333
        )
        
        titreLabel.pack()
        resultatLabel.place(x=0, y=220)

        joueurLabel.place(x=60,y=80)
        scoreJoueurLabel.place(x=100,y=120)
        ordinateurLabel.place(x=160,y=80)
        scoreOrdinateurLabel.place(x=200,y=120)

        quitterButton.place(x=135, y=280)
        resetButton.place(x=210, y=280)        
        pierreButton.place(x=50,y=180)
        feuilleButton.place(x=135,y=180)
        ciseauxButton.place(x=220,y=180)
```
